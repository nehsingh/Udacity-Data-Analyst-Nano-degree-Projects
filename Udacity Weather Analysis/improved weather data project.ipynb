```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading data frame
weather_data = pd.read_csv('Global_city_temp_list.csv')

# analysing and cleaning the dataframe
weather_data = weather_data.fillna(weather_data.mean())
weather_data.info()
```

```python
# getting the rolling averages

output = weather_data.rolling(window = 100 , center = False , on = "city_avg_temp").mean()
```

```python
# ploting the bar graph
plt.figure(figsize = (10,10))
sns.swarmplot(output['year'], output['global_avg_temp'])
plt.xlabel('year')
plt.ylabel('global avg temp')
plt.title('Globl Average Temprature By year')
plt.show()
```

```python
plt.figure(figsize = (10 , 5))
sns.boxplot(output['year'], output['city_avg_temp'])
sns.boxplot(output['year'], output['global_avg_temp'])
plt.xlabel('City Avg Temp')
plt.ylabel('Global Avg Temp')
plt.title('City Avg Temp Vs Global Avg Temp')
plt.show()
```
